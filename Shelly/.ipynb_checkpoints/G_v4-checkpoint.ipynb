```python
import os
import base64
import re
import pandas as pd
from openai import OpenAI

# 🔑 憑證與大會端點（保持妳原本的設定）
GOLDEN_API_KEY = "sk-rc-qLyAj14iCtkrZu9FZuwgYA"
TEAM_NAME = "Zinalrothorn"

client_cscs = OpenAI(
    base_url="https://api.swissai.svc.cscs.ch/v1",
    api_key=GOLDEN_API_KEY,
)

def encode_image_from_renku(image_path: str) -> str:
    """在 Renku 雲端環境讀取本地圖表並轉為 base64"""
    # 判斷副檔名
    if image_path.lower().endswith('.svg'):
        # 雲端如果沒有安裝 cairosvg，建議圖表盡量以 png/jpg 提取
        # 這裡針對 png/jpg 做標準處理
        mime = "image/svg+xml"
    elif image_path.lower().endswith('.jpg') or image_path.lower().endswith('.jpeg'):
        mime = "image/jpeg"
    else:
        mime = "image/png"
        
    with open(image_path, "rb") as image_file:
        data = base64.b64encode(image_file.read()).decode("utf-8")
    return f"data:{mime};base64,{data}"

def ask_with_image_renku(prompt: str, image_path: str, model: str = "moonshotai/Kimi-K2.5-SDSC") -> str:
    """將題目與 Renku 裡的圖表一同餵給 Kimi 視覺模型"""
    try:
        encoded_image = encode_image_from_renku(image_path)
        response = client_cscs.chat.completions.create(
            model=model,
            messages=[
                {
                    "role": "user",
                    "content": [
                        {"type": "text", "text": prompt},
                        {
                            "type": "image_url",
                            "image_url": {"url": encoded_image}
                        },
                    ],
                }
            ],
            temperature=0.1 # 看圖數數字，降低隨機性
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"視覺辨識小失敗: {e}"

print("✅ Renku 專用視覺工具包配置完成！")
```
