````python
# CELL 1: 設定
import os, base64, json, time
from pathlib import Path
from openai import OpenAI
import openpyxl
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter

GOLDEN_API_KEY = 'sk-rc-qLyAj14iCtkrZu9FZuwgYA'
client = OpenAI(
    base_url='https://api.swissai.svc.cscs.ch/v1',
    api_key=GOLDEN_API_KEY,
)

VISION_MODEL   = 'moonshotai/Kimi-K2.5-SDSC'  # ← 若失敗再換這行
IMAGE_FOLDER   = 'extracted_pure_charts'           # ← 你的圖片資料夾
OUTPUT_EXCEL   = 'chart_descriptions.xlsx'

print('✅ 設定完成')
imgs = sorted(Path(IMAGE_FOLDER).glob('*'))
imgs = [p for p in imgs if p.suffix.lower() in {'.png','.jpg','.jpeg','.webp'}]
print(f'📂 找到 {len(imgs)} 張圖片')
for p in imgs[:5]:
    print(f'   {p.name}')
````

````python
# 測試 cell：先跑這個確認模型
test_resp = client.chat.completions.create(
    model=VISION_MODEL,
    messages=[{'role': 'user', 'content': 'Reply with just the word: working'}]
)
print('model reply:', test_resp.choices[0].message.content)

# 再看看有哪些模型可用
models = client.models.list()
vision_candidates = [m.id for m in models.data if any(k in m.id.lower() for k in ['kimi','vl','vision','qwen'])]
print('\n可用的視覺相關模型:')
for m in vision_candidates:
    print(' ', m)
````

````python
# CELL 2: 呼叫 Vision API 描述每張圖

PROMPT = '''Analyze this chart from a research PDF. Return ONLY raw JSON, no markdown fences:
{
  "chart_type": "line chart / bar chart / table / pie chart / infographic / other",
  "title": "chart title or null",
  "main_finding": "most important insight in 1 sentence",
  "data_description": "axes, units, time range, series",
  "key_numbers": ["specific values you can read"],
  "trend": "increasing / decreasing / mixed / stable / N/A",
  "keywords": ["3-5 search keywords"]
}'''

def describe(img_path):
    ext = img_path.suffix.lower().replace('.jpg', '.jpeg').replace('.', '')
    media_type = f'image/{ext}'
    b64 = base64.b64encode(img_path.read_bytes()).decode()
    encoded = f'data:{media_type};base64,{b64}'
    try:
        resp = client.chat.completions.create(
            model=VISION_MODEL,
            messages=[{'role': 'user', 'content': [
                {'type': 'text', 'text': PROMPT},
                {'type': 'image_url', 'image_url': {'url': encoded}}
            ]}],
            temperature=0.5,
        )
        content = resp.choices[0].message.content
        return content if content is not None else 'ERROR: empty response'
    except Exception as e:
        return f'ERROR: {e}'

# processing 

def parse(raw):
    try:
        return json.loads(raw.strip().replace('```json','').replace('```','').strip()), True
    except:
        return raw, False

results = []
for i, img_path in enumerate(imgs):
    print(f'[{i+1:3d}/{len(imgs)}] {img_path.name} ... ', end='')
    raw = describe(img_path)
    parsed, ok = parse(raw)
    
    if i == 0 and raw.startswith('ERROR'):
        print(f'\n\n⛔ 錯誤: {raw[:300]}')
        print('💡 請把錯誤訊息貼給 Claude，換一個支援視覺的模型名稱')
        break
    
    print('✅' if ok else '⚠️ 非JSON')
    results.append({'filename': img_path.name, 'raw': raw, 'parsed': parsed, 'ok': ok})
    time.sleep(0.3)

print(f'\n完成：{sum(r["ok"] for r in results)}/{len(results)} 成功解析為 JSON')
````

````python
print(f'results 共 {len(results)} 筆')
print(f'成功 JSON：{sum(r["ok"] for r in results)} 筆')
print(f'第一筆範例：{results[0]["filename"]} → ok={results[0]["ok"]}')
````

````python
# CELL 3: output to Excel

wb = openpyxl.Workbook()

# ── 樣式 ──
hdr_fill = PatternFill('solid', start_color='1F4E79')
hdr_font = Font(bold=True, color='FFFFFF', name='Arial', size=10)
ok_fill  = PatternFill('solid', start_color='E2EFDA')
err_fill = PatternFill('solid', start_color='FCE4D6')
wrap     = Alignment(wrap_text=True, vertical='top')
center   = Alignment(horizontal='center', vertical='center')
thin     = Border(*[Side(style='thin')]*0,
                  left=Side(style='thin'), right=Side(style='thin'),
                  top=Side(style='thin'),  bottom=Side(style='thin'))

def style_header(ws, headers):
    ws.append(headers)
    for c in range(1, len(headers)+1):
        cell = ws.cell(1, c)
        cell.font, cell.fill, cell.alignment, cell.border = hdr_font, hdr_fill, center, thin
    ws.row_dimensions[1].height = 28

def g(parsed, key):
    if not isinstance(parsed, dict): return ''
    v = parsed.get(key, '')
    return '\n'.join(v) if isinstance(v, list) else (str(v) if v else '')

# ── Sheet 1: 圖表總覽 ──
ws1 = wb.active
ws1.title = '圖表總覽'
cols1 = ['#','檔名','圖表類型','標題','主要發現','資料描述','關鍵數值','趨勢','關鍵字','狀態']
style_header(ws1, cols1)

for i, r in enumerate(results, 1):
    p = r['parsed']
    row = [
        i, r['filename'],
        g(p,'chart_type'), g(p,'title'), g(p,'main_finding'),
        g(p,'data_description'), g(p,'key_numbers'),
        g(p,'trend'), g(p,'keywords'),
        '✅' if r['ok'] else '❌'
    ]
    ws1.append(row)
    row_idx = i + 1
    for c in range(1, 11):
        cell = ws1.cell(row_idx, c)
        cell.font = Font(name='Arial', size=9)
        cell.alignment = wrap
        cell.border = thin
        if c == 10:
            cell.fill = ok_fill if r['ok'] else err_fill
    ws1.row_dimensions[row_idx].height = 55

for col, w in zip('ABCDEFGHIJ', [5,32,16,28,45,45,28,12,28,8]):
    ws1.column_dimensions[col].width = w
ws1.freeze_panes = 'A2'
ws1.auto_filter.ref = f'A1:J1'

# ── Sheet 2: RAG Chunks ──
ws2 = wb.create_sheet('RAG Chunks')
cols2 = ['chunk_id','filename','content_type','chunk_text']
style_header(ws2, cols2)

for i, r in enumerate(results):
    if not r['ok']: continue
    p = r['parsed']
    chunk = (
        f"[CHART] {g(p,'title') or 'Untitled'}\n"
        f"Type: {g(p,'chart_type')}\n"
        f"File: {r['filename']}\n"
        f"Main finding: {g(p,'main_finding')}\n"
        f"Data: {g(p,'data_description')}\n"
        f"Key numbers: {g(p,'key_numbers').replace(chr(10), ' | ')}\n"
        f"Keywords: {g(p,'keywords').replace(chr(10), ', ')}"
    )
    ws2.append([f'chart_{i+1:04d}', r['filename'], 'chart_description', chunk])
    row_idx = ws2.max_row
    for c in range(1,5):
        cell = ws2.cell(row_idx, c)
        cell.font = Font(name='Arial', size=9)
        cell.alignment = wrap
        cell.border = thin
    ws2.row_dimensions[row_idx].height = 80

for col, w in zip('ABCD', [12, 32, 18, 90]):
    ws2.column_dimensions[col].width = w
ws2.freeze_panes = 'A2'

wb.save(OUTPUT_EXCEL)
print(f'✅ Save successfully：{OUTPUT_EXCEL}')
print(f'   Sheet 1「image overview」: {len(results)} 列')
print(f'   Sheet 2「RAG Chunks」: {sum(r["ok"] for r in results)} 列（JSON 成功的）')
````
