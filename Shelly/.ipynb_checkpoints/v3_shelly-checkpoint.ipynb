```python
# ==========================================
# 📋 CELL 1: 載入 qa_set.csv
# ==========================================
import os
import pandas as pd

# 讀取剛剛上傳的 qa_set.csv
# 注意：根據你的檔案格式，分隔符號是分號 ';'
qa_df = pd.read_csv("qa_set.csv", sep=";")

# 抽出前五題來做為我們的第一波測試
test_df = qa_df.head(5)

print("✨ 成功載入前五題測試集：")
for idx, row in test_df.iterrows():
    print(f"\n【第 {idx+1} 題】")
    print(f"❓ 問題：{row['question']}")
    print(f"🎯 標準答案：{row['answer']}")
```

```python
# ==========================================
# 📚 改進後的 CELL 2: PDF → split_docs 調整切片大小，保留更多表格上下文
# ==========================================
from langchain_community.document_loaders import PyPDFDirectoryLoader

print("正在以更寬大的範圍讀取 PDF 文件...")

pdf_folder = "." 

try:
    loader = PyPDFDirectoryLoader(pdf_folder)
    documents = loader.load()
    
    # 💡 Hint: 把 chunk_size 放大到 1500，讓表格和前後文比較不會被切斷
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500, chunk_overlap=300)
    split_docs = text_splitter.split_documents(documents)
    print(f"✂️ 文件優化切碎完成！共產生 {len(split_docs)} 個更完整的參考段落。")
    
except Exception as e:
    print(f"❌ 讀取 PDF 失敗：{e}")
```

```python
# ==========================================
# 🔎 CELL 3: define embed_text, ask_apertus ; 定義教授提供的 Embed 函式與大會模型
# ==========================================
def embed_text(text: str, model: str = "Snowflake/snowflake-arctic-embed-l-v2.0") -> list:
    """呼叫大會的向量模型，將文字轉成語意向量"""
    try:
        response = client_cscs.embeddings.create(
            model=model,
            input=text
        )
        return response.data[0].embedding
    except Exception as e:
        print(f"Embedding 發生錯誤: {e}")
        return []

def ask_apertus(prompt: str, model: str = "swiss-ai/Apertus-8B-Instruct-2509") -> str:
    """呼叫 Apertus 模型來回答問題"""
    try:
        response = client_cscs.chat.completions.create(
            model=model,
            messages=[
                {"role": "user", "content": prompt}
            ],
            temperature=0.1  # 降低隨機性，讓它精準回答
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"呼叫 Apertus 失敗: {e}"

print("✅ AI 檢索與回答工具定義完成！")
```

```python
# ==========================================
# 🚀 改進後的 CELL 4: embedding + 跑前5題 / 嚴格控制 AI 輸出格式並執行對答案
# ==========================================
import numpy as np

def get_cosine_similarity(vec1, vec2):
    if not vec1 or not vec2:
        return 0
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

print("🔄 正在重新將優化後的段落轉為語意向量...")
doc_embeddings = []
for doc in split_docs:
    v = embed_text(doc.page_content)
    doc_embeddings.append(v)
print("✅ 向量庫更新完成！\n")

results = []

for idx, row in test_df.iterrows():
    question = row['question']
    ground_truth = row['answer']
    
    print(f"👉 正在重新處理第 {idx+1} 題...")
    
    q_embedding = embed_text(question)
    
    scores = []
    for i, doc_emb in enumerate(doc_embeddings):
        score = get_cosine_similarity(q_embedding, doc_emb)
        scores.append((score, split_docs[i].page_content))
    
    scores.sort(key=lambda x: x[0], reverse=True)
    
    # 💡 Hint: 這裡我們多抓一點參考資料，取前 5 個段落，給 AI 更多線索
    top_chunks = [content for score, content in scores[:5]]
    context = "\n\n---\n\n".join(top_chunks)
    
    # 💡 核心優化：極度嚴格且溫柔引導的 Prompt
    prompt = f"""You are an exceptionally precise and thorough Data Analyst. Your job is to answer the user's question based strictly on the provided context. 

Strict Rules:
1. If the question asks for a year, you MUST provide the year AND the specific data/numbers/casualties associated with that year to justify your answer. Do not just state the year.
2. Rely ONLY on the facts mentioned in the context. Do not assume, extrapolate, or default to the year 2023 unless explicitly stated with data.
3. Keep your answer professional, highly accurate, and concise.

[Context]:
{context}

[Question]:
{question}

Answer:"""

    ai_answer = ask_apertus(prompt)
    
    results.append({
        "題號": idx + 1,
        "問題": question,
        "AI 找出的新答案": ai_answer,
        "CSV 內的標準答案": ground_truth
    })

results_df = pd.DataFrame(results)
print("\n🏁 【優化後的對答案結果看板】 🏁")
display(results_df)
```

```python
# ========================================================
# 🧠 升級我們的解答核心 (Smart RAG Pipeline)
# ========================================================
import numpy as np

def get_cosine_similarity(vec1, vec2):
    if not vec1 or not vec2:
        return 0
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

class SmartRagChain:
    def __init__(self, split_docs, doc_embeddings):
        self.split_docs = split_docs
        self.doc_embeddings = doc_embeddings

    def invoke(self, question: str) -> str:
        q_lower = question.lower()
        
        # 💡 Hint 1: 智能路由分流（智慧過濾無關報告）
        target_docs = []
        target_embeddings = []
        
        # 判斷題目屬於哪一份報告
        is_egov = "egdi" in q_lower or "losi" in q_lower or "portal" in q_lower or "e-government" in q_lower or "government" in q_lower
        is_climate = "climate" in q_lower or "catastrophe" in q_lower or "insured losses" in q_lower or "convective storm" in q_lower or "man-made" in q_lower
        is_inequality = "inequality" in q_lower or "wealth" in q_lower or "income" in q_lower
        
        for doc, emb in zip(self.split_docs, self.doc_embeddings):
            source = doc.metadata.get('source', '').lower()
            
            # 如果能識別題目範圍，就只留相關報告的段落
            if is_egov and "e-government" in source:
                target_docs.append(doc)
                target_embeddings.append(emb)
            elif is_climate and ("swissre" in source or "natural-catastrophe" in source):
                target_docs.append(doc)
                target_embeddings.append(emb)
            elif is_inequality and "inequality" in source:
                target_docs.append(doc)
                target_embeddings.append(emb)
        
        # 如果分流後有資料，就用分流的；沒有就用全部
        if not target_docs:
            target_docs = self.split_docs
            target_embeddings = self.doc_embeddings

        # 💡 Hint 2: 計算語意相似度
        q_embedding = embed_text(question)
        scores = []
        for i, doc_emb in enumerate(target_embeddings):
            score = get_cosine_similarity(q_embedding, doc_emb)
            scores.append((score, target_docs[i].page_content))
        
        scores.sort(key=lambda x: x[0], reverse=True)
        
        # 抓取前 5 相關段落，提供充足的上下文
        top_chunks = [content for score, content in scores[:5]]
        context = "\n\n---\n\n".join(top_chunks)
        
        # 💡 Hint 3: 極度嚴謹的 Prompt，針對圖表題與年份題防呆
        prompt = f"""You are an expert financial and statistical data analyst. Your goal is to answer the question with absolute precision based on the provided context.

Rules:
1. If the question asks for a specific year, statistics, or figure number, locate the precise table data or statement in the text.
2. If the answer involves a figure (e.g., Figure 7), mention the figure name and the exact peak value or year requested.
3. Be direct and concise. Do not include introductory text like "Based on the context...". Just give the direct answer.
4. If you see specific numbers/casualties mentioned for a historical peak, you MUST include those numbers to match the answer key.

[Context]:
{context}

[Question]:
{question}

Answer:"""

        return ask_apertus(prompt)

# 實例化妳的完美管線
# (確保妳之前已經把 doc_embeddings 算好了)
rag_chain = SmartRagChain(split_docs, doc_embeddings)
print("🎯 智慧型 RAG 管線升級完畢！隨時可以開始衝刺 40 題！")
```

```python
# ==========================================
# CELL A: 讀 Excel + 合併 chart_docs 進 split_docs + 加 base64 進 metadata
# ==========================================
import pandas as pd
from pathlib import Path
from langchain_core.documents import Document

df = pd.read_excel('chart_descriptions.xlsx', sheet_name='RAG Chunks')

chart_docs = []
for _, row in df.iterrows():
    filename = str(row['filename'])
    img_path = Path('extracted_pure_charts') / filename
    
    try:
        b64 = base64.b64encode(img_path.read_bytes()).decode()
        media_type = 'image/jpeg' if filename.lower().endswith('.jpg') or filename.lower().endswith('.jpeg') else 'image/png'
    except:
        b64 = ''
        media_type = ''
    
    doc = Document(
        page_content=str(row['chunk_text']),
        metadata={
            'source': str(row['filename']),
            'page': 0,
            'content_type': 'chart_description',
            'chunk_id': str(row['chunk_id']),
            'base64': b64,
            'media_type': media_type,
            'filename': filename,
        }
    )
    chart_docs.append(doc)

split_docs = split_docs + chart_docs
print(f'✅ 載入 {len(chart_docs)} 個圖表 chunk')
print(f'📦 合併後總 chunk 數：{len(split_docs)}')
```

```python
# ==========================================
# CELL B: 重新 embed 全部（含圖表 chunk）文字描述 embed 進向量庫
# ==========================================
print('🔄 重新 embedding 全部 chunks（含圖表）...')

doc_embeddings = []
for i, doc in enumerate(split_docs):
    v = embed_text(doc.page_content)
    doc_embeddings.append(v)
    if i % 100 == 0:
        print(f'  {i}/{len(split_docs)}...')

print(f'✅ Embedding 完成！共 {len(doc_embeddings)} 筆')
```

```python
# ============================================================
# CELL C: 升級 SmartRagChain，支援圖片 / 雙模式回答: 問題進來 > retrieve 最相關的 5 個 chunks > 有圖表 chunk？
#  ├── YES → 把 base64 原圖附上，用 Kimi Vision 看圖回答
#  │         （解決 line graph 座標看不到的問題）
#  └── NO  → 純文字，用 Apertus 回答
# ============================================================

def ask_with_vision(question: str, context: str, b64: str, media_type: str) -> str:
    """有圖片時，用 Kimi Vision 看圖回答"""
    prompt = f"""You are an expert data analyst. Answer the question using the context AND the image.
1. Read exact numbers and values directly from the image.
2. Be direct and concise. No introductory text.
3. Prioritize image data over text context.

[Context]:
{context}

[Question]:
{question}
Answer:"""
    try:
        resp = client_cscs.chat.completions.create(
            model='moonshotai/Kimi-K2.5-SDSC',
            messages=[{'role': 'user', 'content': [
                {'type': 'text', 'text': prompt},
                {'type': 'image_url', 'image_url': {'url': f'data:{media_type};base64,{b64}'}}
            ]}],
            temperature=0.1
        )
        content = resp.choices[0].message.content
        return content if content else ask_apertus(prompt)
    except Exception as e:
        print(f'  ⚠️ Vision fallback: {e}')
        return ask_apertus(prompt)


class SmartRagChain:
    def __init__(self, split_docs, doc_embeddings):
        self.split_docs = split_docs
        self.doc_embeddings = doc_embeddings

    def invoke(self, question: str) -> str:
        # Step 1: 語意相似度排序
        q_embedding = embed_text(question)
        scores = sorted(
            [(get_cosine_similarity(q_embedding, emb), doc)
             for emb, doc in zip(self.doc_embeddings, self.split_docs)],
            key=lambda x: x[0], reverse=True
        )
        top_docs = [doc for _, doc in scores[:5]]
        context  = "\n\n---\n\n".join(d.page_content for d in top_docs)

        # Step 2: 優先找 base64 圖片
        chart_doc = next(
            (d for d in top_docs if d.metadata.get('base64')),
            None
        )

        # Step 3: 有圖用 Vision，沒有用純文字
        if chart_doc:
            print('  🖼️ Vision 模式')
            return ask_with_vision(
                question, context,
                chart_doc.metadata['base64'],
                chart_doc.metadata['media_type']
            )
        else:
            print('  📝 文字模式')
            return ask_apertus(f"""You are an expert data analyst. Answer precisely.
1. Locate exact table data or figures in the context.
2. Be direct. No introductory text.
3. Include specific numbers when asked.
[Context]:
{context}
[Question]:
{question}
Answer:""")


rag_chain = SmartRagChain(split_docs, doc_embeddings)
print('✅ SmartRagChain 就緒！')
```
