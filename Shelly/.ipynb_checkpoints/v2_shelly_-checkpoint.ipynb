```python
pdf_folder = "../File" 
pdf_files = []

if os.path.exists(pdf_folder):
    for file in os.listdir(pdf_folder):
        if file.endswith(".pdf"):
            pdf_files.append(os.path.join(pdf_folder, file))
else:
    pdf_folder = "File"
    if os.path.exists(pdf_folder):
        for file in os.listdir(pdf_folder):
            if file.endswith(".pdf"):
                pdf_files.append(os.path.join(pdf_folder, file))

raw_docs = []
for file_path in pdf_files:
    print(f"📖 正在載入: {file_path}")
    loader = PyPDFLoader(file_path)
    raw_docs.extend(loader.load())

# ✨ 細細加工：補全縮寫，防止語意丟失
for doc in raw_docs:
    text = doc.page_content
    text = text.replace("Nat Cat", "Natural Catastrophe")
    text = text.replace("Nat-Cat", "Natural Catastrophe")
    text = text.replace("man-made disasters", "man-made catastrophes")
    doc.page_content = text

# 切碎文件
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1200, chunk_overlap=250)
splits = text_splitter.split_documents(raw_docs) 
print(f"🎉 成功完成語意修補，並切碎成 {len(splits)} 個文字方塊！")
```

```python
print("📦 正在下載並調用本地 HuggingFace Embedding 將文字向量化...")
print("💡 (第一次執行會花大約 30 秒下載模型，請稍等它一下喔～)")

embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-small-en-v1.5",
    model_kwargs={'device': 'cpu'}
)

vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 20})

print("✅ ")
```

```python
def apertus_brain_wrapper(prompt_input):
    if not isinstance(prompt_input, str):
        prompt_input = str(prompt_input)
    try:
        response = client_cscs.chat.completions.create(
            model="swiss-ai/Apertus-8B-Instruct-2509",  # 教授指定的黃金模型名稱
            messages=[{"role": "user", "content": prompt_input}],
            temperature=0.0  # 保證回答嚴謹，不瞎編數據
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Error during Apertus call: {e}"

print("success")
```

```python
system_prompt = (
    "You are an expert financial and climate data analyst for the Durham Hackathon.\n"
    "Your job is to provide exact data points from the provided context.\n"
    "CRITICAL GUIDELINES:\n"
    "1. Strictly distinguish between 'Natural Catastrophes' (earthquakes, floods) and 'Man-made Disasters' (aviation, industrial).\n"
    "2. Do not blindly guess or infer that a recent year (like 2023) is the answer unless explicit data points back it up.\n"
    "3. Check every year mentioned in the context (including historical years like 2002, 2005, 2011) and compare their numbers strictly.\n"
    "4. Extract the exact percentages or numbers asked. Look carefully at whether the text mentions '11%' or '21%' for different loss metrics.\n\n"
    "【Context】:\n{context}"
)
prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{input}"),
])

def format_docs(documents):
    return "\n\n".join(doc.page_content for doc in documents)

# 終極 RAG Pipeline
rag_chain = (
    {"context": retriever | format_docs, "input": RunnablePassthrough()}
    | prompt
    | apertus_brain_wrapper  
)
print("🔗 RAG 完美管線建立成功！")
```

```python
BASE_URL = "http://durham-leaderboard-runai-innovation-klemen.inference.compute.datascience.ch"
leaderboard_endpoint = f"{BASE_URL}/api/v1/leaderboard"
submit_endpoint = f"{BASE_URL}/api/v1/submit"


def get_leaderboard() -> pd.DataFrame:
    leaderboard = requests.get(leaderboard_endpoint).json()
    return pd.DataFrame(leaderboard["entries"])


def submit(df: pd.DataFrame, user: str, token: str) -> pd.Series:
    if user == "your_team_name" or user == "":
        raise ValueError("Please set your team name in the 'TEAM_NAME' variable.")
    predictions = df[["question", "answer"]].to_dict(orient="records")
    submission = {"predictions": predictions}
    response = requests.post(submit_endpoint, json=submission, auth=(user, token))
    if (response.status_code // 100) != 2:
        response.raise_for_status()
    print(response.json())
    return pd.Series(response.json())
```

```python
import os
import requests
import pandas as pd
from openai import OpenAI
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough

# 1. 套件安全檢測
try:
    from langchain_huggingface import HuggingFaceEmbeddings
except ModuleNotFoundError:
    from langchain_community.embeddings import HuggingFaceEmbeddings
try:
    from langchain_chroma import Chroma
except ModuleNotFoundError:
    from langchain_community.vectorstores import Chroma
```

```python
# ========================================================
# 🔑 設定憑證與大會端點
# ========================================================
GOLDEN_API_KEY = "sk-rc-qLyAj14iCtkrZu9FZuwgYA"
TEAM_NAME = "Zinalrothorn"
BASE_URL = "http://durham-leaderboard-runai-innovation-klemen.inference.compute.datascience.ch"
submit_endpoint = f"{BASE_URL}/api/v1/submit"

client_cscs = OpenAI(
    base_url="https://api.swissai.svc.cscs.ch/v1",
    api_key=GOLDEN_API_KEY,
)
```

```python

# ========================================================
# 📖 讀取 PDF、優化文字切碎重疊度
# ========================================================
pdf_folder = "../File" if os.path.exists("../File") else "File"
pdf_files = [os.path.join(pdf_folder, f) for f in os.listdir(pdf_folder) if f.endswith(".pdf")]

raw_docs = []
for file_path in pdf_files:
    loader = PyPDFLoader(file_path)
    raw_docs.extend(loader.load())

for doc in raw_docs:
    text = doc.page_content
    text = text.replace("Nat Cat", "Natural Catastrophe").replace("Nat-Cat", "Natural Catastrophe")
    text = text.replace("man-made disasters", "man-made catastrophes")
    doc.page_content = text

# 💡 優化：加大 overlap 到 400，防止跨頁表格數據斷裂
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1200, chunk_overlap=400)
splits = text_splitter.split_documents(raw_docs)

embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en-v1.5", model_kwargs={'device': 'cpu'})
vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 20})
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough

# 💡 帶上妳極簡、嚴格防禦的 System Prompt
system_prompt = (
    "You are a precise QA assistant for a Hackathon scoring system.\n"
    "Your output will be judged by an exact match algorithm. Therefore, keep your answer as CONCISE and SHORT as possible.\n"
    "DO NOT write long full sentences. DO NOT say 'Based on the context...' or 'According to...'.\n"
    "Just directly state the specific year, number, percentage, or name asked (e.g., instead of 'The year was 2011 with 6 events', just write '2011 with 6.' or '2011').\n"
    "天災(Natural Catastrophes)與人為(Man-made)數據必須嚴格分開計算。\n\n"
    "【Context】:\n{context}"
)
prompt = ChatPromptTemplate.from_messages([("system", system_prompt), ("human", "{input}")])
format_docs = lambda docs: "\n\n".join(d.page_content for d in docs)

# 完美組裝管線
rag_chain = (
    {"context": retriever | format_docs, "input": RunnablePassthrough()} 
    | prompt 
    | apertus_brain_wrapper
)
print("🔗 RAG 完美管線建立成功！")
```

```python
# ========================================================
# 🚀 【接續步驟】將 rag_chain 丟進 40 題測試並自動提交
# ========================================================

# 1. 讀取大會的原始題目 CSV 檔（後面 35 題是空白的）
csv_path = "../../Durham-Hackathon-2026-template/data/a2q/qa_set.csv"
if not os.path.exists(csv_path): 
    csv_path = "qa_set.csv"  # 防呆機制

df_qa = pd.read_csv(csv_path, encoding="utf-8", sep=";").fillna("")
print(f"📋 題目集加載成功！總共包含 {len(df_qa)} 道題目。")
print("🔥 Zinalrothorn 隊的全自動答題引擎，正式開始丟進測試囉...")

# 2. 開始跑 40 題的迴圈測試
answered_count = 0

for index, row in df_qa.iterrows():
    # 💡 關鍵過濾網：只針對答案是空白的（第 6 到 40 題）丟進 RAG 測試
    if row['answer'] == "": 
        question = row['question']
        print(f"\n🤖 [測試中] 正在讓 rag_chain 解答第 {index+1}/{len(df_qa)} 題...")
        
        try:
            # 傳送門 🚪：把題目丟進妳做好的 rag_chain 完美管線裡！
            answer = rag_chain.invoke(question)
            clean_answer = answer.strip()
            
            # 將 RAG 測試出來的答案精準填回 Dataframe
            df_qa.at[index, 'answer'] = clean_answer
            print(f"💡 RAG 回答成功: {clean_answer[:120]}...")
            answered_count += 1
            
        except Exception as e:
            print(f"❌ 第 {index+1} 題發生小意外: {e}")
            df_qa.at[index, 'answer'] = "I don't know"

print(f"\n🎉 精彩絕倫！35 道測試題目已經全部被 rag_chain 攻克填補完畢！")

# ========================================================
# 🏆 3. 自動把寫滿答案的成果丟進大會官方 submit 系統
# ========================================================
print(f"\n🚀 正在將 Zinalrothorn 隊的最終戰果提交至大會官方端點...")
try:
    submit_result = submit(df_qa, user=TEAM_NAME, token="cant_be_empty")
    print("\n🏆 ✨ 萬歲！！！全自動測試與排行榜提交成功！快去網頁看看你們跳到第幾名了！ ✨ 🏆")
except Exception as e:
    print(f"⚠️ 提交給大會官方時發生網路小卡頓: {e}")
    backup_file = "final_submission_zinalrothorn.csv"
    df_qa.to_csv(backup_file, index=False, sep=";")
    print(f"💾 別慌！已經幫妳把寫滿珍貴答案的成果存檔至本地目錄：'{backup_file}'")
    
```

```python
# ========================================================
# 📸 Zinalrothorn 隊：純 Python PDF 逐頁闪电圖片化工具 (100% 繞過 poppler)
# ========================================================
import os
import glob

# 1. 🛡️ 安全安裝純 Python 的高級 PDF 渲染利器 (無痛、免系統補丁)
print("📦 正在安全配置本機 PyMuPDF 圖片轉換元件...")
os.system("pip install --upgrade pip")
os.system("pip install PyMuPDF")

import fitz  # PyMuPDF 的官方導入名稱

# 2. 🎯 設定輸入與指定的輸出資料夾名稱
pdf_folder = "../File"
output_dir = "./pdf_image"  # 完美對齊妳指定的資料夾名稱！

if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"📁 成功建立全新的圖片儲存載體：'{output_dir}'")

# 3. 🔍 自動點名原始 PDF 檔案
pdf_paths = glob.glob(os.path.join(pdf_folder, "*.pdf"))
print(f"🕵️‍♂️ 成功鎖定 {len(pdf_paths)} 本即將被圖片化的黃金大報告...")

# 4. 🚀 點火執行！逐本、逐頁將 PDF 渲染成高解析度 PNG 圖片
for pdf_path in pdf_paths:
    pdf_name = os.path.basename(pdf_path)
    pdf_name_clean = os.path.splitext(pdf_name)[0]
    print(f"\n⏳ 正在將報告閃電圖片化: {pdf_name} ...")
    
    try:
        # 打開 PDF 檔案
        doc = fitz.open(pdf_path)
        total_pages = len(doc)
        print(f"  📊 偵測成功！這本報告總共有 {total_pages} 頁")
        
        # 逐頁進行高品質渲染
        for page_idx in range(total_pages):
            page = doc.load_page(page_idx)
            
            # 💡 關鍵細節：設定縮放矩陣以大幅提升圖片清晰度 (150 DPI 等級，看圖表不模糊)
            zoom = 2.0  
            mat = fitz.Matrix(zoom, zoom)
            pix = page.get_pixmap(matrix=mat)
            
            # 檔名格式整整齊齊：如 swissre_sigma_1_page_001.png
            page_num = page_idx + 1
            output_filename = f"{pdf_name_clean}_page_{page_num:03d}.png"
            output_filepath = os.path.join(output_dir, output_filename)
            
            # 安全儲存為 PNG
            pix.save(output_filepath)
            
        print(f"  ✅ {pdf_name} 每一頁都已成功轉化為獨立漂亮的 PNG 圖片！")
        
    except Exception as e:
        print(f"  ❌ 處理 {pdf_name} 時發生小卡頓: {e}")

print(f"\n🎉 萬歲！！！Zinalrothorn 隊多模態圖片資產 100% 建置成功！")
print(f"🚀 快去左邊的瀏覽器看看，幾百張完美的頁面圖片已經乖乖躺在 '{output_dir}' 裡面囉！")
```

```python
import os
import signal
import pandas as pd
import pdfplumber
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

extra_docs = []

# ============================================================
# 1. 圖表描述 txt（你朋友已生成的）
# ============================================================
caption_dir = "/home/renku/work/Durham-Hackathon-2026-w1t2/Sandy/image_captions_kimi_final"
for txt_file in os.listdir(caption_dir):
    if not txt_file.endswith(".txt"):
        continue
    with open(os.path.join(caption_dir, txt_file), "r", encoding="utf-8") as f:
        content = f.read()
    extra_docs.append(Document(
        page_content=f"[VISUAL CHART DATA - {txt_file}]\n{content}",
        metadata={"source": txt_file, "type": "chart_caption"}
    ))
print(f"📊 圖表描述：{len(extra_docs)} 個")
```

```python

# ============================================================
# 2. 表格 + 腳註（用 extract_tables 不用 find_tables，不會卡死）
# ============================================================
pdf_folder = "../File"
table_count = 0
footnote_count = 0

for pdf_file in sorted(os.listdir(pdf_folder)):
    if not pdf_file.endswith(".pdf"):
        continue
    pdf_path = os.path.join(pdf_folder, pdf_file)
    print(f"\n📖 處理：{pdf_file}")

    try:
        with pdfplumber.open(pdf_path) as pdf:
            last_header = None
            last_col_count = 0

            for page_num, page in enumerate(pdf.pages):
                # --- 腳註：抽頁面底部 18% ---
                try:
                    footnote_box = page.within_bbox((0, page.height * 0.82, page.width, page.height))
                    footnote_text = footnote_box.extract_text()
                    if footnote_text and len(footnote_text.strip()) > 20:
                        extra_docs.append(Document(
                            page_content=f"[FOOTNOTE - {pdf_file} page {page_num+1}]\n{footnote_text.strip()}",
                            metadata={"source": pdf_file, "page": page_num+1, "type": "footnote"}
                        ))
                        footnote_count += 1
                except:
                    pass

                # --- 表格：用 extract_tables 比 find_tables 穩定 ---
                try:
                    tables = page.extract_tables()
                except:
                    continue
                if not tables:
                    continue

                for t_idx, table in enumerate(tables):
                    if not table:
                        continue
                    clean = [[(c.strip() if c else "") for c in row] for row in table if any(row)]
                    if len(clean) < 2:
                        continue

                    current_col_count = len(clean[0])

                    # 跨頁表格焊接
                    if t_idx == 0 and last_header and current_col_count == last_col_count:
                        if clean[0] != last_header:
                            clean.insert(0, last_header)

                    try:
                        df = pd.DataFrame(clean[1:], columns=clean[0])
                        content = df.to_markdown(index=False)
                    except:
                        content = "\n".join([" | ".join(row) for row in clean])

                    last_header = clean[0]
                    last_col_count = current_col_count

                    extra_docs.append(Document(
                        page_content=f"[TABLE - {pdf_file} page {page_num+1}]\n{content}",
                        metadata={"source": pdf_file, "page": page_num+1, "type": "table"}
                    ))
                    table_count += 1

    except Exception as e:
        print(f"  ❌ {pdf_file}: {e}")

print(f"\n📋 表格：{table_count} 個")
print(f"📝 腳註：{footnote_count} 個")
print(f"📦 總共：{len(extra_docs)} 個額外文件")
```

```python
# ============================================================
# 3. 切碎並加進你原本的 vectorstore
# ============================================================
splitter = RecursiveCharacterTextSplitter(chunk_size=1200, chunk_overlap=200)
extra_splits = splitter.split_documents(extra_docs)
print(f"✂️ 切碎後：{len(extra_splits)} 個 chunks")

vectorstore.add_documents(extra_splits)
retriever = vectorstore.as_retriever(search_kwargs={"k": 25})
print("✅ vectorstore 更新完成！現在包含文字 + 表格 + 腳註 + 圖表描述")
```

```python
# 重建 vectorstore（只需要跑這一段）
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
try:
    from langchain_huggingface import HuggingFaceEmbeddings
except:
    from langchain_community.embeddings import HuggingFaceEmbeddings
try:
    from langchain_chroma import Chroma
except:
    from langchain_community.vectorstores import Chroma

pdf_folder = "../File"
raw_docs = []
for f in sorted(os.listdir(pdf_folder)):
    if f.endswith(".pdf"):
        loader = PyPDFLoader(os.path.join(pdf_folder, f))
        raw_docs.extend(loader.load())

for doc in raw_docs:
    doc.page_content = doc.page_content.replace("Nat Cat", "Natural Catastrophe").replace("Nat-Cat", "Natural Catastrophe")

splits = RecursiveCharacterTextSplitter(chunk_size=1200, chunk_overlap=400).split_documents(raw_docs)
print(f"📄 原始文字：{len(splits)} chunks")

# 合併 extra_splits（表格+腳註+圖表，剛才已切好）
all_splits = splits + extra_splits
print(f"📦 總共：{len(all_splits)} chunks")

embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en-v1.5", model_kwargs={'device': 'cpu'})
vectorstore = Chroma.from_documents(documents=all_splits, embedding=embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 25})
print("✅ vectorstore 建立完成！")
```
