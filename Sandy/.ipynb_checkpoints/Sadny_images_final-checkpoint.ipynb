```python
import os

clean_caption_dir = "/home/renku/work/Durham-Hackathon-2026-w1t2/Sandy/image_captions_kimi_final"

sandy_chart_chunks = []

if not os.path.exists(clean_caption_dir):
    print("⚠️ 錯誤：找不到 Sandy 的 image_captions_kimi_final 資料夾，請確認路徑！")
else:
    txt_files = [f for f in os.listdir(clean_caption_dir) if f.endswith('.txt')]
    print(f"🕵️‍♂️ 正在載入 Sandy 組的核心圖表文字庫，共計 {len(txt_files)} 個高密度數據說明書...")
    
    for txt_file in txt_files:
        file_path = os.path.join(clean_caption_dir, txt_file)
        with open(file_path, "r", encoding="utf-8") as f:
            content = f.read()
        
        # 建立標準 RAG 區塊結構，並打上強標籤
        sandy_chart_chunks.append({
            "page_content": f"[CRITICAL VISUAL DATA CHART - SOURCE: {txt_file}]\n{content}",
            "metadata": {
                "source": "vlm_chart_caption",
                "original_file": txt_file,
                "data_type": "visual_chart"
            }
        })

    print(f"\n✅ 【圖表組交貨成功！】")
    print(f"📊 封裝完成的圖表 Chunks 數量：{len(sandy_chart_chunks)} 個")
    print(f"🚀 變數名稱：`sandy_chart_chunks` 已在記憶體中就位，隨時可以與文字組、表格組進行 `.extend()` 合併！")
```
